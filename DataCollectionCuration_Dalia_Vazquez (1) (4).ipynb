```python
# we are renaming the columns names to more easy terms to work with

gdf.columns = ['geoid',
 'name',
 'Total',
 'Income in the past 12 months below poverty level:',
 'Male Total:',
 'Male 18 to 24 years',
 'Male 25 to 34 years',
 'Male 35 to 44 years',
 'Male 45 to 54 years',
 'Male 55 to 64 years',
 'Male 65 to 74 years',
 'Male 75 years and over',
 'Female Total:',
 'Female 18 to 24 years',
 'Female 25 to 34 years',
 'Female 35 to 44 years',
 'Female 45 to 54 years',
 'Female 55 to 64 years',
 'Female 65 to 74 years',
 'Female 75 years and over',
 'geometry']
```

```python
gdf.head() # double check the names were updated
```

```python
gdf['Female Total:'].head() # Lets look at the first 5 values for female total, shows us the type of data we are working with
```

```python
gdf_sorted = gdf.sort_values(by='Female Total:',ascending = False) # ascend the Female Total column
```

```python
gdf_sorted[['name','Female Total:']].head(10) # get the top 10 counties with the highest female pop below poverty
```

```python
gdf_sorted.head(10).plot(figsize=(10,10)) # lets plot the results from above. This will only provide us with the coordinates of the top 10 counties, it is showing where they are located
```

```python
# Demonstrate all 254 counties and the female total below the poverty rate
# not very useful or accurate because we are just diplaying the total and comparing the rate of the total amongst all women below the poverty level
# the current female total includes children which makes this analysis inaccurate

gdf_sorted.head(254).plot(figsize=(10,10),
                          column='Female Total:', 
                          cmap='inferno', 
                          legend=True)
```

```python
# we need to update the male and female totals to only include the adults and not the children

gdf['Female Total:'] =   gdf['Female 18 to 24 years'] + gdf['Female 25 to 34 years'] + gdf['Female 35 to 44 years'] + gdf['Female 45 to 54 years'] + gdf['Female 55 to 64 years'] + gdf['Female 65 to 74 years'] + gdf['Female 75 years and over']
gdf['Male Total:'] =   gdf['Male 18 to 24 years'] + gdf['Male 25 to 34 years'] + gdf['Male 35 to 44 years'] + gdf['Male 45 to 54 years'] + gdf['Male 55 to 64 years'] + gdf['Male 65 to 74 years'] + gdf['Male 75 years and over']
```

```python
gdf.sample(5) # get a sample to make sure that the total add up and are updated
```

```python
gdf['Total'] = gdf['Male Total:'] + gdf['Female Total:'] # update the total column by adding the new and updated male and female totals
```

```python
gdf.sample(5) # get a sample ensure it was updated
```

```python
# create a new column 
#normalize the data by gender 
# we can do this by using the female total and dividing the total pop below poverty and multiply by 100 to get the percent
gdf['Percent of Latinas in Poverty'] = gdf['Female Total:'] / gdf['Total'] * 100
```

```python
gdf.sample(5) # make sure the column is present
```

```python
#create a new column
#normalize the data by gender 
# we can do this by using the male total and dividing the total pop below poverty and multiply by 100 to get the percent

gdf['Percent of Latinos in Poverty'] = gdf['Male Total:'] / gdf['Total'] * 100
```

```python
gdf.sample(5)  # make sure column is present
```

```python
# we will create a 1x2 subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 12))
ax1, ax2 = axs

# percent of Latinas below poverty rate 
gdf.plot(column='Percent of Latinas in Poverty', #column you want to use
            cmap='inferno', #color scheme 
            scheme='naturalbreaks',# what type of break do you want
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax1, # what axis are we assining it to 
            legend=True) # add legend

ax1.axis("off")
ax1.set_title("Percent of Latinas Living Below Poverty (Past 12 Months)")

# percent of Latinos below poverty rate
gdf.plot(column='Percent of Latinos in Poverty', #column you want to use 
            cmap='inferno', #color scheme
            scheme='naturalbreaks', # what type of break do you want
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax2, # what axis are we assining it to 
            legend=True)# add legend
            
             
           

ax2.axis("off")
ax2.set_title("Percent of Latinos Living Below Poverty (Past 12 Months) ")
```

The map visualization above represents the Male and Female percent of the total Latino population living below the poverty level for every county in Texas.
