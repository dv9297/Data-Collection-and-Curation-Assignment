```python
gdf['Female Total:'] =   gdf['Female 18 to 24 years'] + gdf['Female 25 to 34 years'] + gdf['Female35 to 44 years'] + gdf['Female 45 to 54 years'] + gdf['Female 55 to 64 years'] + gdf['Female 65 to 74 years'] + gdf['Female 75 years and over']
gdf['Male Total:'] =   gdf['Male 18 to 24 years'] + gdf['Male 25 to 34 years'] + gdf['Male 35 to 44 years'] + gdf['Male 45 to 54 years'] + gdf['Male 55 to 64 years'] + gdf['Male65 to 74 years'] + gdf['Male 75 years and over']
```

```python
gdf.sample(5)
```

```python
gdf['Total'] = gdf['Male Total:'] + gdf['Female Total:']
```

```python
gdf.sample(5)
```

```python
list(gdf)
```

```python
gdf['Percent of Latinas in Poverty'] = gdf['Female Total:'] / gdf['Total'] * 100
```

```python
gdf.sample(5)
```

```python
gdf['Percent of Latinos in Poverty'] = gdf['Male Total:'] / gdf['Total'] * 100
```

```python
gdf.sample(5)
```

```python
# create the 1x2 subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 12))

# name each subplot
ax1, ax2 = axs

# percent hispanic map on the left
gdf.plot(column='Percent of Latinas in Poverty', 
            cmap='inferno', 
            scheme='naturalbreaks',
            classification_kwds={'bins':[20,40,60,80,100]},
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax1, # this assigns the map to the subplot,
            legend=True
           )

ax1.axis("off")
ax1.set_title("Percent of Latinas Living Below Poverty (Past 12 Months) ")

# percent black map on the right
gdf.plot(column='Percent of Latinos in Poverty', 
            cmap='inferno', 
            scheme='naturalbreaks',
            classification_kwds={'bins':[20,40,60,80,100]},
            edgecolor='white', 
            linewidth=0., 
            alpha=0.75, 
            ax=ax2, # this assigns the map to the subplot
            legend=True
           )

ax2.axis("off")
ax2.set_title("Percent of Latinos 45 to 54 years Living Below Poverty (Past 12 Months) ")
```

The map visualization above represents the Male and Female percent of the total Latino population living below the poverty level for every county in Texas.
